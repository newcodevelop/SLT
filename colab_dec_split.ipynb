```
import pickle
import gzip

def load_dataset_file(filename):
  with gzip.open(filename, "rb") as f:
    loaded_object = pickle.load(f)
    return loaded_object

p = load_dataset_file('phoenix14t.pami0.train')

names = []
for i in tqdm(range(len(p))):
  names.append(p[i]['name'])
```

```
names[0]
```

```
names = list(map(lambda x: 'span=12_stride=2/'+x.split('/')[1]+'.pt',names))
```

```
names[0]
```

```
os.path.join('train',names[0].split('/')[1]).split('.')[0]
```

```
import tensorflow as tf
import numpy as np
os.makedirs('train')
l = []
ref = np.random.randn(238,1024)
for i in tqdm(names):
  path = os.path.join('train',i.split('/')[1]).split('.')[0]

  val = np.asarray(torch.squeeze(torch.stack(torch.load(i)),dim=1),dtype = np.float32)
  l.append(val.shape[0])
  sequence = [val,ref]
  k = tf.keras.preprocessing.sequence.pad_sequences(sequence, padding='post', dtype = 'float32')
  np.save(str(path),k[0,:,:])
```
